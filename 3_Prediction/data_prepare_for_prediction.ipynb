```python
# -*- coding: utf-8 -*-
import pandas as pd
from datetime import datetime


dfhourly=pd.read_csv(r'C:\luyan\paper_code\data\hourly.csv')

dfhourly["state"]=0
dfhourly.loc[dfhourly["con"]>0,"state"]=1


dfhourly["time_stamp"]=dfhourly["day_id"]
dfhourly['hour']=dfhourly["time_stamp"].astype(str).str[8:10]

dfhourly['weekday']=pd.to_datetime(dfhourly['time_stamp'],format="%Y%m%d%H").apply(datetime.weekday)


dfhourly['daytype']='work'
dfhourly.loc[(dfhourly['weekday']==5)|(dfhourly['weekday']==6),'daytype']='rest'

dfclimate=pd.read_csv(r'C:\luyan\paper_code\data\cq2016_climate.csv')

dfhourly=pd.merge(dfhourly,dfclimate,on="time_stamp")


dfhourly=dfhourly[['id','daytype','hour','tout','state']]

dfwm_thermal_label=pd.read_csv(r'C:\luyan\paper_code\data\thermal_cluster_label_wm.csv')
dfwm_schedule_label=pd.read_csv(r'C:\luyan\paper_code\data\schedule_cluster_label_wm.csv')
dfwm_label=pd.merge(dfwm_thermal_label,dfwm_schedule_label,on='id')[['id','thermal_type','schedule_type']]

dfhourly_wm=pd.merge(dfhourly,dfwm_label,on='id')

dfhourly_wm=dfhourly_wm[['id','daytype','hour','tout','thermal_type','schedule_type','state']]

dffs_thermal_label=pd.read_csv(r'C:\luyan\paper_code\data\thermal_cluster_label_fs.csv')
dffs_schedule_label=pd.read_csv(r'C:\luyan\paper_code\data\schedule_cluster_label_fs.csv')
dffs_label=pd.merge(dffs_thermal_label,dffs_schedule_label,on='id')[['id','thermal_type','schedule_type']]

dfhourly_fs=pd.merge(dfhourly,dffs_label,on='id')

dfhourly_fs=dfhourly_fs[['id','daytype','hour','tout','thermal_type','schedule_type','state']]

#dfhourly_wm.to_csv(r'C:\luyan\paper_code\data\hourly_for_pre_wm.csv',index=False)
#dfhourly_fs.to_csv(r'C:\luyan\paper_code\data\hourly_for_pre_fs.csv',index=False)
dfhourly_wm
dfhourly_fs
```
